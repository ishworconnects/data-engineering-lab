```python
#Create  data Frame and add REVENUE column to it
import pandas as pd

data = {
    "product": ["phone", "macbook", "watch", "airpods"],
    "price": [1000, 1500, 500, 200],
    "sales": [10, 12, 15, 22]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Add a new column revenue

df["revenue"] = df["price"] * df["sales"]
print(df)
```

```python
#add a record to the data
#add new record using LOC-to locate row
#to return one or more specified rows

df.loc[len(df)] = ["laptop", 1000,13,1000*13]
print(df)
```

```python
#Calculate total revenue

Total_revenue = df["revenue"].sum()
print(Total_revenue)
```

```python
#add record KeyBOARD using loc


df.loc[len(df)] = ["Charger", 200,11,200*11]
print(df)
```

```python
#to see index value 0 and 1

print(df.loc[[0,1]])

```

```python
#Missing Values
```

```python
df.loc[2,"sales"]=None
print(df)
```

```python
#fill missing values
#fillna - allows a replace empty sales with values
df["sales"].fillna(0, inplace=True)
print(df)
```

```python
#Filter High Value products
high_value = df[df["price"] > 13000]
print(high_value)
```

```python
#sort columns
sorted_df = df.sort_values(by="revenue", ascending=False)
print(sorted_df)
```

```python
import pandas as pd

data = {
    "Name": ["Ram", "Shyam","Ishwor"],
    "Age": [25, 30, 22],
    "City": ["Kathmandu", None, "Lalitpur"]
}

df = pd.DataFrame(data)
print(df)
```

```python
#detect the missing values
print(df.isnull())
#isnull returns True if value is missing and False if value is present
```

```python
#Count the missing value
print(df.isnull().sum())
```

```python
#fill missing values- fillna

df.fillna(0,inplace=True)
print(df)
```

```python
import pandas as pd

# create data
dbs = {
    "product": ["macbook", "laptop", "iphone", "mouse"],
    "sales": [10, None, 14, 15],
    "revenue": [1000, 2000, 1200, 1000]
}

# convert into DataFrame
df = pd.DataFrame(dbs)

# fill missing values with mean average
df["sales"].fillna(df["sales"].mean(), inplace=True)

# print DataFrame
print(df)
```

```python
# create data
dabs = {
    "product": ["macbook", "laptop", "iphone", "mouse"],
    "sales": [10, None, 14, 15],
    "revenue": [1000, 2000, 1200, 1000]
}

# convert into DataFrame
df = pd.DataFrame(dabs)

# remove missing values
df = df.dropna()

# print DataFrame
print(df)
```
