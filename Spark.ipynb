from pyspark.sql import SparkSession

```python
%python
#Creating raw data
data= [
    (1,"phone",20000),
    (2,"laptop",30000),
    (3,"TV",40000),
    (4,"Remote",500)

]
columns = ["order_id","Products","Price"]

df = spark.createDataFrame(data,columns)
display(df)

#appply filter and Select
result=df.filter(df.Price>20000).select("Products","Price")
display(result)
```

```python
from pyspark.sql.functions import sum

data1 = [
    (1, "TV", 20000),
    (2, "laptop", 30000),
    (3, "TV", 40000),
    (4, "Remote", 500)
]

columns1 = ["order_id1", "Products1", "Price1"]

df = spark.createDataFrame(data1, columns1)

display(df)

result = df.groupBy("Products1") \
           .agg(sum("Price1").alias("Total_Sales"))

display(result)
```

```python
#Xreate any new column - add a new column discount_price = price-100
from pyspark.sql.functions import col #col access column

df= spark.createDataFrame(data1,columns1)

new_result = df.withColumn("discount_price", col("Price1")-100)
display(new_result)
```

```python
#to remove duplicates

drop_df = df.dropDuplicates(["Products1"])
display(drop_df)
```
