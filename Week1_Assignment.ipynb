Question 1: Explain the difference between List and Tuple.
--> List and Tuple are both collections used to store multiple items in Python. Lists = changeable/mutable, Tuples = fixed/immutable

Also write a Python program to:
    1. Create a list
	2.Convert it into a tuple
	3.Try modifying both


```python
# Creating a list
my_list = [1, 2, 3, 4, 5]
print("List:", my_list)
```

```python
#Converting it into a tuple
my_Tuple = tuple(my_list)
print("Tuple:" , my_Tuple)
```

```python
# Modifying the list
my_list[0] = 100
print("Modifying list:", my_list)
```

```python
#TRY Modifying Tuple
my_Tuple[0] = 15
print(my_Tuple) # cannot be modified
```

```python
try:
    my_Tuple[0] = 99
except TypeError:
    print("Tuple cannot be modified")
```

Question 2: What is a Set?
---> A set is a collection that stores unique values only. It does not allow duplicate values and does not maintain order. It is usually created using curly braces {}
my_set = {1, 2, 3, 4}

Explain how sets remove duplicates.
--->Sets remove duplicates automatically because they store only unique values. When duplicate items are added, the set keeps only one copy and ignores the repeated values.

Write a program to remove duplicates from a list and print unique values.


```python
#Removing duplicates using set
numbers = [1, 2, 2, 3, 4, 4, 5,8,8,9]
unique_numbers = set(numbers)

print("Original Numbers:", numbers)
print("Unique Numbers ", unique_numbers)
```

Question 3: Write a program to:
		Reverse a list using slicing
		Also print original list to show it is unchanged
 

```python
my_list = [5, 7, 3, 4, 10]
reversed_list = my_list[::-1]

print("Original List:", my_list)
print("Reversed List:", reversed_list)
```

Question 4: Write a program to:
	1.	Create a dictionary of student details
	2.	Access the name
	3.	Remove age using pop()
	4.	Print length of dictionary


```python
student = {
    "name": "Ishwor",
    "age": 32,
    "course": "Data Engineering Python"
}

print("My Dictionary: " , student)
print("Name:", student["name"]) #Accessing Name

print("Length:", len(student))

student.pop("age") # Removing Age using pop()
print("After removing Age:" , student)

#Length of Dictionary after removing
print("Length:", len(student))
```
