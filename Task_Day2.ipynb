#Create a List

```python
my_list = [10,20,30]

#Find the length of the list

print("Original list :",my_list)
print("Length of list :",len(my_list))
```

```python
#Add 2 elements to the list
my_list.append(40)
my_list.append(50)  

#Display the list after adding elements
print("List after adding elements :",my_list)   
```

Tuple

```python
#Create a tuple
my_tuple = (1,2,3,4,5,6)
#Find the length of the tuple
print("Original tuple :",my_tuple)
print("Length of tuple :",len(my_tuple))
```

```python
#tuple slicing
#Perform slicing to get a subset
subset = my_tuple[2:5]  # This will get elements from index 2 to 4
print("Subset of tuple (index 2 to 4) :",subset)
```

Using Zip() with Two Lists

```python
#Create two lists
list1 = [1,2,3]
list2 = ['a' ,'b' ,'c'] 

#Using zip() and convert to a list of tuples
zipped_list = list(zip(list1, list2))

#Display result
print("Zipped list of tuples :",zipped_list)
```
