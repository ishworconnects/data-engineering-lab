```python
roll_no = {1, 2, 3, 4, 5}
print(roll_no)  
```

```python
#Adding item to set
event_types = {"click", "view", "purchase"}
event_types.add("signup") # Adding item to the set
print(event_types)
```

```python
#Removing item from set
event_types = {"click", "view", "purchase", "signup"}
event_types.remove("signup") # Removing item from the set
print(event_types)
```

```python
event_types = {"click", "view", "purchase"}
event_types.pop() # Removing an arbitrary item from the set
print(event_types)
```

```python
#Check membership in set
event_types = {"click", "view", "purchase"}
print("click" in event_types) # Checking if "click" is in the set
print("signup" in event_types) # Checking if "signup" is in the set
```

```python
#Remove duplicate items from list
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = set(numbers) # Converting list to set to remove duplicates

print(unique_numbers)
```

```python
customer = {
    "id": 1,
    "name": "John Doe",
    "email": "john.doe@example.com",
    "city": "Helsinki"
}
print(customer)
```

```python
#Accessing dictionary values
customer = {
    "id": 1,
    "name": "John Doe",
    "email": "john.doe@example.com",
    "city": "Helsinki"
}
print(customer["name"])  # Accessing the "name" value
print(customer["email"])  # Accessing the "email" value
print(customer["city"])  # Accessing the "city" value
```

```python
#Change dictionary values
customer ["city"] = "Espoo" # Modifying the "city" value
print(customer)
```

```python
#Add new key-value pair to dictionary
customer["age"] = 30 # Adding new key-value pair to the dictionary
print(customer) 
```

```python
#Remove key-value pair from dictionary
del customer["age"] # Removing key-value pair from the dictionary
print(customer)
```

```python
customer.pop("name") # Removing key-value pair using pop method
print(customer)
```

```python
#Loop and dictionary
customer = {
    "id": 1,
    "name": "John Doe",
    "email": "john.doe@example.com",
    "city": "Helsinki"
}
for key, value in customer.items():
    print(key, value)
```

```python
#Safe access with get method
customer = { "name" : "Ishwor"}
print(customer.get("name")) # Accessing existing key
print(customer.get("email")) # Accessing non-existing key   
print(customer.get("email", "No email provided")) # Accessing non-existing key with default value
```
