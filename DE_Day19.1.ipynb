```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
orders = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
        .load("/Volumes/workspace/de_spark/datasets/orders_csv.csv")
```

```python
users =spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
        .load("/Volumes/workspace/de_spark/datasets/customers_csv.csv")
```

```python
new_orders= spark.read.format("json") \
    .load("/Volumes/workspace/de_spark/datasets/new_orders.json")
```

spark.sql("query")

```python
orders.createOrReplaceTempView("orders")
users.createOrReplaceTempView("users")

#it will convert the dataFrame --> into sql table(it is temporary(temp))
```

```python
result=spark.sql("""
    select *
    from orders
    where amount > 300
    """
)
result.show()
```

```python
users.show()
```

```python
#join wuiith aggregation
revenue_df=spark.sql("""
                     select 
                     u.country,
                     count(o.order_id) as total_orders,
                     sum(o.amount) as total_revenue
                     from orders o
                     join users u
                     on o.customer_id=u.customer_id
                     group by u.country
                     order by total_revenue desc
                     """)
revenue_df.show()
                     
```

```python
min_amount=300
filter_df=spark.sql(f"""
                   select *
                   from orders
                   where amount > {min_amount}  
                    """

)
```

```python
#we have to update and insert
#find new records with the help of join

new_orders=spark.sql("""
                     select *
                     from new_orders n
                     left join orders o
                     on n.order_id=o.order_id
                     where o.order_id is null
                     """)
```

```python
new_orders.show()
```

```python
#updated record
updated_record=spark.sql("""
                     select *
                     from new_orders n
                     join orders o
                     on n.order_id=o.order_id
                     """)
```
