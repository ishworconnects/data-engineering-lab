Create a variable temperature = 32. Write a program that prints "Hot" if the temperature is greater than 30, otherwise print "Normal"

```python
temperature = 32

if temperature > 30:
    print("Hot")
else:
    print("Normal")
```

Write a function called add_tax(price) that adds 13% tax to a product price and returns the final price. Test it with 100

```python
def add_tax(price):
    tax = price * 0.13
    final_price = price + tax
    return final_price

# Testing with 100
result = add_tax(100)
print(f"Original price: $100")
print(f"Final price with 13% tax: ${result}")
```

Given this list:
expenses = [200, None, 450, 100, None, 300]

Use a loop to print only the values that are not None.

```python
expenses = [200, None, 450, 100, None, 300]

for expense in expenses:
    if expense is not None:
        print(expense)
```

Create a class called Car with brand and year. Make one object using your favorite car brand and print both values.

```python
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

# Creating an object with a car brand
my_car = Car("Tesla", 2024)

# Printing both values
print(f"Brand: {my_car.brand}")
print(f"Year: {my_car.year}")
```
